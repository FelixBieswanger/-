```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def mittel(data):
    return sum(data)/len(data)

def varianz(data, mittelwert):
    s = 0
    
    for x in data:
        s+=pow((x-mittelwert),2)
    
    return (1/(len(data)-1))*s

def pearson(x_list,y_list,x_mittel,y_mittel,s_x,s_y):
    temp = 0
    
    for x,y in zip(x_list,y_list):
        temp += ((x-x_mittel)*(y-y_mittel))
        
    temp *= (1/(len(x_list)-1))
    
    return temp/(s_x*s_y)

def regression(x,y):
    x_mittel = sum(x)/len(x)
    y_mittel = sum(y)/len(y)

    s2_x = varianz(x,x_mittel)
    s_x = math.sqrt(s2_x)
    print("standardabweichung x",round(s_x,3))
    s2_y = varianz(y,y_mittel)
    s_y = math.sqrt(s2_y)
    print("standardabweichung y",round(s_y,3))
    
    r_xy = pearson(x,y,x_mittel,y_mittel,s_x,s_y)
    print("pearson",round(r_xy,4))
    b = r_xy*(s_y/s_x)
    a = y_mittel - b*x_mittel
    print("y= ",round(a,3),"+",round(b,3),"x")
    
    x_l = np.linspace(min(x),max(x),len(x))
    y_l = b*x_l+a

    plt.figure(figsize=(13,6))
    plt.title("gesucht y=a' + b' *x")
    plt.xlabel("x",fontsize=25)
    plt.ylabel("y",fontsize=25)
    plt.scatter(x,y,marker="o",c="black")
    plt.plot(x_l,y_l,c="r")
    plt.show()
```

```python
x = [0.1, 0.5, 0.8, 1, 1.2, 1.4, 1.6, 2, 2.2]
y = [0.07, 0.15, 0.19, 0.23, 0.33, 0.35, 0.4, 0.51, 0.5]
regression(x,y)
```

```python
x2 = [0.2,0.5 ,0.8, 1 ,1.2, 1.4, 1.6, 2, 2.2, 2.4, 2.7 ,3]
y2 = [0.16,0.13, 0.14, 0.14, 0.11, 0.12, 0.11, 0.12, 0.1, 0.14, 0.13 ,0.16]
regression(x2,y2)
```
